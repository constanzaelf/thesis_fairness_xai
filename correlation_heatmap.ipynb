```python
import numpy as np
from scipy.stats import pearsonr, spearmanr
from sklearn.metrics import jaccard_score
from scipy.stats import entropy
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import cosine
from scipy.special import kl_div
import matplotlib.pyplot as plt

def pearson_correlation(matrix1, matrix2): # the higher the value the more similar are the images
    flattened1 = matrix1.flatten()
    flattened2 = matrix2.flatten()
    return pearsonr(flattened1, flattened2)[0]

def spearman_correlation(matrix1, matrix2): # the higher the value the more similar are the images
    flattened1 = matrix1.flatten()
    flattened2 = matrix2.flatten()
    return spearmanr(flattened1, flattened2)[0]

def jaccard_similarity(matrix1, matrix2):
    flattened1 = matrix1.flatten().tolist()
    flattened2 = matrix2.flatten().tolist()
    return jaccard_score(flattened1, flattened2)

def kl_divergence(matrix1, matrix2): #the higher the value the greater the difference between the images 
    flattened1 = matrix1.flatten()
    flattened2 = matrix2.flatten()
    return entropy(flattened1, flattened2)

def cosine_distance(matrix1, matrix2): # the lower the value the more similar are the images
    flattened1 = matrix1.flatten()
    flattened2 = matrix2.flatten()
    
    return cosine([flattened1], [flattened2])
```

```python
hm_all_fp = np.load('matrix_values/total_matrix_false_positives.npy')
hm_male_fp = np.load('matrix_values/total_matrix_false_positives_male.npy')
hm_female_fp = np.load('matrix_values/total_matrix_false_positives_female.npy')

# Similarities calculation
pearson_corr = pearson_correlation(hm_female_fp, hm_male_fp)
spearman_corr = spearman_correlation(hm_female_fp, hm_male_fp)
kl_div = kl_divergence(hm_female_fp, hm_male_fp)
cosine_dist = cosine_distance(hm_female_fp, hm_male_fp)
print("False Positives:\n")
print("Correlation Pearson Male:", pearson_corr)
print("Correlation Spearman Male:", spearman_corr)
#print("Coeficiente de Similaridad de Jaccard:", jaccard_sim)
print("Divergence Kullback-Leibler Male:", kl_div)
print("Cosinus Distance Male:", cosine_dist)

hm_all_fp = np.load('matrix_values/total_matrix_false_positives.npy')
hm_male_fp_pp = np.load('matrix_values/total_matrix_false_positives_pp_male.npy')
hm_female_fp_pp = np.load('matrix_values/total_matrix_false_positives_pp_fem.npy')

# Calculando las medidas de similitud
pearson_corr_pp = pearson_correlation(hm_female_fp_pp, hm_male_fp_pp)
spearman_corr_pp = spearman_correlation(hm_female_fp_pp, hm_male_fp_pp)
#jaccard_sim = jaccard_similarity(hm_all_fp, hm_male_fp)
kl_div_pp = kl_divergence(hm_female_fp_pp, hm_male_fp_pp)
cosine_dist_pp = cosine_distance(hm_female_fp_pp, hm_male_fp_pp)
print("False Positives:\n")
print("Correlation Pearson Male:", pearson_corr_pp)
print("Correlation Spearman Male:", spearman_corr_pp)
print("Divergence Kullback-Leibler Male:", kl_div_pp)
print("Cosinus Distance Male:", cosine_dist_pp)


```
