```python
import csv

archivo_entrada = 'parsed_files/celeba-att-male-partitions.csv'
archivo_salida = 'parsed_files/resultado.csv'

lineas_filtradas = []

with open(archivo_entrada, 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    for linea in csv_reader:
        if linea[2] == '2' and linea[3] == '1':
            lineas_filtradas.append(linea)

with open(archivo_salida, 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerows(lineas_filtradas)

```
